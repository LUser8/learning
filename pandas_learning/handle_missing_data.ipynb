```python
# handle missing data in pandas using fillna, interpolate and dropna methods
import pandas as pd
```

```python
df = pd.read_csv('stock_data.csv')
```

```python
df
```

```python
df = pd.read_csv('weather_data.csv', parse_dates=['day'])
df.set_index('day', inplace=True)
```

```python
df
```

```python
# Fillna
df.fillna(0)
```

```python
df.fillna({'temperature': 0, 'windspeed': 0, 'event': 'no event'})
```

```python
df.fillna(method='ffill')
```

```python
df.ffill()
```

```python
df.bfill()
```

```python
df.interpolate()
```

```python
df.infer_objects(copy=False)
```

```python
df.dropna()
```

```python
df.dropna(how='all')
```

```python
df.dropna(thresh=2)
```

```python
# to replace specific values in dataframe
df = pd.read_csv('weather_data.csv', parse_dates=['day'])
```

```python
df
```

```python
df.replace(-99999, 0)
```

```python
import numpy as np
df.replace(['0', -99999, -88888], np.NaN)
```

```python
df = df.replace({'temperature':[-99999, -88888], 'windspeed': '-88888', 'event': '0'}, np.NaN)
```

```python
df
```

```python
df.replace({'temperature':['[A-Za-z]'], 'windspeed': '[A-Za-z]', 'event': '0'}, '', regex=True)
```
