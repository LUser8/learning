```python
# group by and concat df
import pandas as pd
df = pd.read_csv('weather_data_by_cities.csv')
df
```

```python
g = df.groupby('city')
g
```

```python
for city, city_df in g:
    print(city)
    print(city_df)
```

```python
g.get_group('mumbai')
```

```python
g.max()
```

```python
g.mean()
```

```python
g.describe()
```

```python
india_weather = {
    'city': ['mumbai', 'delhi', 'bangalore'],
    'temperature': [21, 14, 35],
    'humidity': [68, 65, 72]
}
india_w_df = pd.DataFrame(india_weather)
india_w_df
```

```python
us_weather = {
    'city': ['newyork', 'chicago', 'orlando'],
    'temperature': [21, 14, 35],
    'humidity': [68, 65, 72]
}
us_w_df = pd.DataFrame(us_weather)
us_w_df
```

```python
df = pd.concat([india_w_df, us_w_df])
df
```

```python
df = pd.concat([india_w_df, us_w_df], ignore_index=True)
df
```

```python
df = pd.concat([india_w_df, us_w_df], ignore_index=True, keys=['india', 'us'])
df
```
